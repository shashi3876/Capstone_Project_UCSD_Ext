```python

file_path = '/content/gdrive/My Drive/Capstone Project/Data/tweets.csv'
 
tweets_btc = pd.read_csv(file_path,sep=';')
print(len(tweets_btc))
tweets_btc=tweets_btc.drop(columns=['id','url'])
tweets_btc.dropna(inplace=True)
tweets_btc. isnull(). sum() * 100 / len(tweets_btc)
print(len(tweets_btc))
save_path = '/content/gdrive/My Drive/Capstone Project/Data/tweets_preprocessed.csv'
with open(save_path, 'w') as f:
  tweets_btc.to_csv(f)

```

```python
tweets_btc['timestamp1'] = pd.to_datetime(tweets_btc['timestamp'])
tweets_btc['date'] = tweets_btc['timestamp1'].dt.date
tweets_btc['time'] = tweets_btc['timestamp1'].dt.time
tweets_btc['zone'] = tweets_btc['timestamp1'].dt.tz

import datetime
cutoff=datetime.datetime.strptime('20:00:00','%H:%M:%S').time()

tweets_btc_new = tweets_btc[tweets_btc['time']<cutoff]
tweets_btc_new=tweets_btc_new.drop(columns=['timestamp1','timestamp','zone'])
print(tweets_btc_new.sample(20))

print(len(tweets_btc_new))

save_path = '/content/gdrive/My Drive/Capstone Project/Data/tweets_preprocessed_before8.csv'
with open(save_path, 'w') as f:
  tweets_btc_new.to_csv(f)
```
