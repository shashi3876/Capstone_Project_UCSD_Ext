## Handling null values

Data set seems to have a lot of missing values. 
Lets see what percentage of valeus are null for each column.

```python
tweets_btc. isnull(). sum() * 100 / len(tweets_btc)
```

Drop the whole url column and id column.

```python
tweets_btc=tweets_btc.drop(columns=['id','url'])
tweets_btc. isnull(). sum() * 100 / len(tweets_btc)
```

```python
tweets_btc.dropna(inplace=True)
tweets_btc. isnull(). sum() * 100 / len(tweets_btc)
```

# Remove non English tweets

When we look at the dataset, some of the entries are in other languages. In this section we'll try to remove non English tweets so that we can correctly use the sentiment analysis packages and other NLP later.

```python
!pip install langdetect
from langdetect import detect

def is_en(txt):
  try:
    return detect(txt)=='en'
  except:
    return False

#tweets_btc_english = tweets_btc[tweets_btc['text'].apply(is_en)]
```

```python
tweets_btc_english = tweets_btc[tweets_btc['text'].apply(is_en)]
tweets_btc_english.info()
```

# Saving the preprocessed dataset

Now we have a preprocessed data set with no null values, extracting useful columns, and only selecting English tweets. Save the new data frame to a csv file in the drive.


```python
save_path = '/content/gdrive/My Drive/Capstone Project/Data/tweets_processed_en_only.csv'
with open(save_path, 'w') as f:
  tweets_btc_english.to_csv(f)
```
