## Some data cleaning

We'll clean the text part of the tweets.


```python
import re

def clean(raw):
    """ Remove hyperlinks and markup """
    result = re.sub("<[a][^>]*>(.+?)</[a]>", 'Link.', raw)
    result = re.sub('&gt;', "", result)
    result = re.sub('&#x27;', "'", result)
    result = re.sub('&quot;', '"', result)
    result = re.sub('&#x2F;', ' ', result)
    result = re.sub('<p>', ' ', result)
    result = re.sub('</i>', '', result)
    result = re.sub('&#62;', '', result)
    result = re.sub('<i>', ' ', result)
    result = re.sub("\n", '', result)
    return result
```

```python
tweets_btc['cleaned'] = tweets_btc.text.parallel_apply(clean)
```

## Use of TextBlob for sentiment analysis

First lets setup TextBlob and check whether it works in a simple sentence.



```python
!pip install textblob
from textblob import TextBlob

sent="Everyone is very sad"
answer=TextBlob(sent)
print(answer.sentiment)

sent="We are really excited"
answer=TextBlob(sent)
print(answer.sentiment)
```

Now we'll use this sentiment analysis on the text column of our data set and add the sentiment to the tweet dataset.

```python
def getSubjectivity(text):
  return TextBlob(text).sentiment.subjectivity
  
def getPolarity(text):
  return TextBlob(text).sentiment.polarity
```

```python
%%time

tweets_btc['tb_subjectivity']= tweets_btc['cleaned'].parallel_apply(getSubjectivity) 
tweets_btc['tb_polarity']= tweets_btc['cleaned'].parallel_apply(getPolarity) 
tweet_part=tweets_btc[['cleaned','tb_polarity']]
tweet_part.sample(10)
```

## Adding VADER sentiment values

```python
!pip install nltk
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sent_i = SentimentIntensityAnalyzer()
```

```python
%%time

#tweets_btc=tweets_btc.sample(frac=0.1,random_state=42)
tweets_btc['vd_sentiment'] = [sent_i.polarity_scores(v)['compound'] for v in tweets_btc['cleaned']]
tweets_btc[['cleaned','tb_polarity','vd_sentiment']].head(20)
```

## Getting the average sentiment for each day

Now since we hae three diffretn sentimen values from three different methods, we'll aggregate these data for each day. Also we'll get the total number of tweets for each day. At the end, we will get a dataframe with date as the index. This will be combined with the BTC price dataframe.

```python
tweets_btc.info()
```

```python
#print(tweets_btc.sample(100))
tweets_btc.info()
tweets_btc.dropna(inplace=True)
print(tweets_btc.head(10))
tweets_btc['retweets'] = pd.to_numeric(tweets_btc['retweets'], errors='coerce')#.astype('float')
tweets_btc['likes'] = pd.to_numeric(tweets_btc['retweets'], errors='coerce')
tweets_btc['replies'] = pd.to_numeric(tweets_btc['replies'], errors='coerce')
#tweets_btc['likes'] = tweets_btc['likes'].astype('float')
#tweets_btc['replies'] = tweets_btc['replies'].astype('float')
tweets_btc.info()
print(tweets_btc.head(10))

save_path = 'Data/tweets_sentiment_added_final_full_sample.csv'
with open(save_path, 'w') as f:
    tweets_btc.to_csv(f)
    
    
grp = tweets_btc.groupby('date')
 
tweets_btc_sentiments=grp.agg({'text' : 'count', 'likes' : 'mean','retweets' : 'mean','replies' : 'mean','tb_polarity' : 'mean','tb_subjectivity' : 'mean','vd_sentiment' : 'mean'})
tweets_btc_sentiments.head(10)
```
